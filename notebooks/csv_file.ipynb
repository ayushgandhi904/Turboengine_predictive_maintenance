```python
import pandas as pd
```

```python
df_train1 = pd.read_csv("./data/train_FD001.txt", sep = " ", header= None).drop(columns=[26,27])
df_test1 = pd.read_csv("./data/test_FD001.txt", sep = " ", header = None).drop(columns=[26,27])
df_RUL1 = pd.read_csv("./data/RUL_FD001.txt", sep=" ", header = None).drop(columns=[1])


df_train2 = pd.read_csv("./data/train_FD002.txt", sep = " ", header= None).drop(columns=[26,27])
df_test2 = pd.read_csv("./data/test_FD002.txt", sep = " ", header = None).drop(columns=[26,27])
df_RUL2 = pd.read_csv("./data/RUL_FD002.txt", sep=" ", header = None).drop(columns=[1])


df_train3 = pd.read_csv("./data/train_FD003.txt", sep = " ", header= None).drop(columns=[26,27])
df_test3 = pd.read_csv("./data/test_FD003.txt", sep = " ", header = None).drop(columns=[26,27])
df_RUL3 = pd.read_csv("./data/RUL_FD003.txt", sep=" ", header = None).drop(columns=[1])


df_train4 = pd.read_csv("./data/train_FD004.txt", sep = " ", header= None).drop(columns=[26,27])
df_test4 = pd.read_csv("./data/test_FD004.txt", sep = " ", header = None).drop(columns=[26,27])
df_RUL4 = pd.read_csv("./data/RUL_FD004.txt", sep=" ", header = None).drop(columns=[1])
```

```python
new_column_names = [
    "Engine_no",
    "Cycle_no",
    "Setting_1",
    "Setting_2",
    "Setting_3",
    "Fan_inlet_temperature",
    "LPC_outlet_temperature",
    "HPC_outlet_temperature",
    "LPT_outlet_temperature",
    "Fan_inlet_pressure",
    "Bypass_duct_pressure",
    "HPC_outlet_pressure",
    "Physical_fan_speed",
    "Physical_core_speed",
    "Engine_pressure_ratio",
    "HPC_outlet_static_pressure",
    "Fuel_flow_ratio",
    "Fan_speed",
    "Core_speed",
    "Bypass_ratio",
    "Burner_fuel_air_ratio",
    "Bleed_enthalpy",
    "Required_fan_speed",
    "Required_fan_conversion_speed",
    "High_pressure_cool_air_flow",
    "Low_pressure_cool_air_flow"
]

rul_cols = ["RUL"]
```

```python
df_train1.columns = new_column_names
df_test1.columns = new_column_names
df_RUL1.columns = rul_cols
df_train2.columns = new_column_names
df_test2.columns = new_column_names
df_RUL2.columns = rul_cols
df_train3.columns = new_column_names
df_test3.columns = new_column_names
df_RUL3.columns = rul_cols
df_train4.columns = new_column_names
df_test4.columns = new_column_names
df_RUL4.columns = rul_cols
```

```python
df_train1['RUL_max'] = df_train1.groupby('Engine_no')['Cycle_no'].transform('max')
df_train1["RUL"] = df_train1["RUL_max"] - df_train1["Cycle_no"]
df_train1.drop("RUL_max", axis = 1, inplace = True)
df_test1['RUL_max'] = df_test1.groupby('Engine_no')['Cycle_no'].transform('max')
df_test1["RUL"] = df_test1["RUL_max"] - df_test1["Cycle_no"]
df_test1.drop("RUL_max", axis = 1, inplace = True)

df_train2['RUL_max'] = df_train2.groupby('Engine_no')['Cycle_no'].transform('max')
df_train2["RUL"] = df_train2["RUL_max"] - df_train2["Cycle_no"]
df_train2.drop("RUL_max", axis = 1, inplace = True)
df_test2['RUL_max'] = df_test2.groupby('Engine_no')['Cycle_no'].transform('max')
df_test2["RUL"] = df_test2["RUL_max"] - df_test2["Cycle_no"]
df_test2.drop("RUL_max", axis = 1, inplace = True)

df_train3['RUL_max'] = df_train3.groupby('Engine_no')['Cycle_no'].transform('max')
df_train3["RUL"] = df_train3["RUL_max"] - df_train3["Cycle_no"]
df_train3.drop("RUL_max", axis = 1, inplace = True)
df_test3['RUL_max'] = df_test3.groupby('Engine_no')['Cycle_no'].transform('max')
df_test3["RUL"] = df_test3["RUL_max"] - df_test3["Cycle_no"]
df_test3.drop("RUL_max", axis = 1, inplace = True)

df_train4['RUL_max'] = df_train4.groupby('Engine_no')['Cycle_no'].transform('max')
df_train4["RUL"] = df_train4["RUL_max"] - df_train4["Cycle_no"]
df_train4.drop("RUL_max", axis = 1, inplace = True)
df_test4['RUL_max'] = df_test4.groupby('Engine_no')['Cycle_no'].transform('max')
df_test4["RUL"] = df_test4["RUL_max"] - df_test4["Cycle_no"]
df_test4.drop("RUL_max", axis = 1, inplace = True)
```

```python
df_RUL1["Engine_no"] = range(1, max(df_test1["Engine_no"])+1)
df_RUL2["Engine_no"] = range(1, max(df_test2["Engine_no"])+1)
df_RUL3["Engine_no"] = range(1, max(df_test3["Engine_no"])+1)
df_RUL4["Engine_no"] = range(1, max(df_test4["Engine_no"])+1)
```

```python
merge_df1 = df_test1.merge(df_RUL1, on = "Engine_no", how = "left")
merge_df1.rename(columns = {"RUL": "more_RUL"}, inplace = True)
merge_df1["RUL"] = merge_df1["RUL_x"] + merge_df1["RUL_y"]
merge_df1.drop(columns= ["RUL_x", "RUL_y"], inplace = True)
df_test1 = merge_df1

merge_df2 = df_test2.merge(df_RUL2, on = "Engine_no", how = "left")
merge_df2.rename(columns = {"RUL": "more_RUL"}, inplace = True)
merge_df2["RUL"] = merge_df2["RUL_x"] + merge_df2["RUL_y"]
merge_df2.drop(columns= ["RUL_x", "RUL_y"], inplace = True)
df_test2 = merge_df2

merge_df3 = df_test3.merge(df_RUL3, on = "Engine_no", how = "left")
merge_df3.rename(columns = {"RUL": "more_RUL"}, inplace = True)
merge_df3["RUL"] = merge_df3["RUL_x"] + merge_df3["RUL_y"]
merge_df3.drop(columns= ["RUL_x", "RUL_y"], inplace = True)
df_test3 = merge_df3

merge_df4 = df_test1.merge(df_RUL4, on = "Engine_no", how = "left")
merge_df4.rename(columns = {"RUL": "more_RUL"}, inplace = True)
merge_df4["RUL"] = merge_df4["RUL_x"] + merge_df4["RUL_y"]
merge_df4.drop(columns= ["RUL_x", "RUL_y"], inplace = True)
df_test4 = merge_df4
```

```python
df_train1.to_csv("train1.csv", index = False)
df_test1.to_csv("test1.csv", index = False)

df_train2.to_csv("train2.csv", index = False)
df_test2.to_csv("test2.csv", index = False)

df_train3.to_csv("train3.csv", index = False)
df_test3.to_csv("test3.csv", index = False)

df_train4.to_csv("train4.csv", index = False)
df_test4.to_csv("test4.csv", index = False)
```

```python
train = {}
test = {}

for i in range(1, 5):
    train[i] = pd.read_csv(f"train{i}.csv")
    test[i] = pd.read_csv(f"test{i}.csv")
```
