```python
columns = {0:"Engine_no",1:"Cycle_no",2:"Setting_1",3:"Setting_2",4:"Setting_3",5:"Fan_inlet_temperature",6:"LPC_outlet_temperature",
           7:"HPC_outlet_temperature",8:"LPT_outlet_temperature",9:"Fan_inlet_pressure",10:"Bypass_duct_pressure",
           11:"HPC_outlet_pressure",12:"Physical_fan_speed",13:"Physical_core_speed",14:"Engine_pressure_ratio",
           15:"HPC_outlet_static_pressure",16:"Fuel_flow_ratio",17:"Fan_speed",18:"Core_speed",19:"Bypass_ratio",20:"Burner_fuel_air_ratio",
           21:"Bleed_enthalpy",22:"Required_fan_speed",23:"Required_fan_conversion_speed",24:"High_pressure_cool_air_flow",25:"Low_pressure_cool_air_flow"}

def getting_df(file_no, columns): 
    df_train = pd.read_csv(f"./data/train_FD00{file_no}.txt" , sep=' ' , header=None).drop(columns=[26,27])
    df_test = pd.read_csv(f"./data/test_FD00{file_no}.txt" , sep=' ' , header=None).drop(columns=[26,27])
    df_RUL = pd.read_csv(f"./data/RUL_FD00{file_no}.txt" , sep=' ' , header=None).drop(columns=[1])

    ## Adding column names : 
    df_train = df_train.rename(columns=columns)
    df_test = df_test.rename(columns=columns)
    df_RUL = df_RUL.rename(columns={0:'RUL'})
    return df_train , df_test , df_RUL 
```
